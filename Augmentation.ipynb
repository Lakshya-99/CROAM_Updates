```
import librosa
import numpy as np
import os
import matplotlib.pyplot as plt
from scipy.io.wavfile import read
from scipy.io.wavfile import write
import shutil
import IPython.display as ipd

!pip install pydub
import pydub
from pydub import AudioSegment
```

```
def manipulate_frequency(data, sampling_rate):
  samples = []
  samples.append(librosa.effects.pitch_shift(data, sampling_rate, n_steps = 0))
  # for i in range(5):
  #   samples.append(librosa.effects.pitch_shift(data, sampling_rate, n_steps = (i+1)))
  #   samples.append(librosa.effects.pitch_shift(data, sampling_rate, n_steps = -1*(i+1)))
  return samples
```

```
def manipulate_speed(data):
  samples = []
  samples.append(librosa.effects.time_stretch(data, 1.25))
  # for i in range(2):
  #   samples.append(librosa.effects.time_stretch(data, (i+5)/4))
  #   samples.append(librosa.effects.time_stretch(data, (i+2)/4))
  return samples
```

```
def match_target_amplitude(sound, target_dBFS):
    change_in_dBFS = target_dBFS - sound.dBFS
    return sound.apply_gain(change_in_dBFS)

# sound = AudioSegment.from_file("yourAudio.m4a", "m4a")
# normalized_sound = match_target_amplitude(sound, -20.0)
# normalized_sound.export("nomrmalizedAudio.m4a", format="mp4")
```

```
#Test an example here

filename = 'Help/Help_Data_Old/1_1.wav'
augmented='new.wav'

fs,x = read(filename)
xs = [float(item) for item in x]
xs=np.asarray(xs)

new = manipulate_speed(xs)

new=np.asarray(new)
write(augmented, fs, new.T)

x, sr = librosa.load(augmented, sr=None, mono=True)
print('Duration: {:.2f}s, {} samples'.format(x.shape[-1] / sr, x.size))
start, end = 0,1
ipd.Audio(data=x[start*sr:end*sr], rate=sr)
```
