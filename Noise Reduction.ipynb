```
#TEST 1
filename = 'new.wav'
x, sr = librosa.load(filename, sr=None, mono=True)
ipd.Audio(data=x, rate=sr)
```

```
noisy_part = x[:]
reduced_noise = nr.reduce_noise(audio_clip=x, noise_clip=noisy_part, verbose=False)
ipd.Audio(reduced_noise, rate=sr)
```
