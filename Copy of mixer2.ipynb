```
import numpy as np
import os
import matplotlib.pyplot as plt
from scipy.io.wavfile import read
from scipy.io.wavfile import write
import shutil
from random import randint
```

```
#Cut a random section of audio
def cut_random_section(noise2, size2):
    size21 = noise2.size
    starting_point2 = randint(0,(noise2.size - size2))
    end_point2 = starting_point2 + size2
    noise_cut_part2 = noise2[starting_point2:end_point2]
    return noise_cut_part2
```

```
def mix(audio1, noise1, snr1, fs1):
   
    #Uncomment only for negative data
    '''
    random_number = int(np.floor(3*fs1*np.random.uniform(0,1)))
    random_starting_number = int(floor(3*fs1*np.random.uniform(0,random_number)))
    audio11 = np.zeros(audio1.shape[0])
    audio11[random_starting_number:random_number] = audio1[random_starting_number:random_number]
    audio1 = audio11
    '''
    audio1 = audio1*1.
    audio1 = audio1/max(audio1)
    noise1 = cut_random_section(noise1, audio1.size)
    noise1 = noise1*1.
    noise1 = noise1/max(noise1)
    gain = pow(10,(snr1/10.))
    numerator = np.mean(abs(audio1)**2)
    denominator = numerator/gain
    noise_power = np.mean(abs(noise1)**2)
    mult_value = (denominator/noise_power)**0.5
    noisy1 = audio1 + noise1*mult_value
    noisy1 = noisy1/max(noisy1)
    return noise1*mult_value, mult_value, noisy1
```
